```python
import json
def add_item_to_todo(uid, id1, title, status):
    new_item = {"userId": uid, "id": id1, "title": title, "completed": status}
    with open('mytodo.json', "r") as file:
        data=json.load(file)
        data.append(new_item)
    with open('mytodo.json', "w") as file:
        json.dump(data, file, indent=4)
        print(f"Task added successfully.")
def display(id1):
    print("For 👤-",id1,"\nTo-Do list contains:\n")
    with open('mytodo.json', 'r') as f:
        data = json.load(f)
        for item in data:
            if item["userId"] == id1:
                print(f"🗒️ {item['title']}","\n")

def todo(id1):
    print("To-Do List: ")
    print("-----------")
    with open('mytodo.json', 'r') as f:
        data = json.load(f)
        for item in data:
            if item["userId"]==id1:
                if "completed" in item and item["completed"] == True:
                    print(f"✔️ {item['title']}\n")
                else:
                    print(f"❌ {item['title']}\n")
'''def changestatus(id1):
    print(" Updated To-Do List: ")
    print("-----------")
    with open('mytodo.json', 'r') as f:
        data = json.load(f)
        for item in data:
            if item["userId"]==id1:
                if "completed" in item and item["completed"] == False:
                    item["completed"]=True
                else:
                    print(f"❌ {item['title']}\n")
'''
def menu():
    print("TO-DO Application")
    print("-----------------")
    while(True):
        print("1. Add items to the list")
        print("2. Display added items for a user")
        print("3. Display the status of the list items")
        print("4. Exit")
        choice=int(input("Enter your choice(1-4): "))
        if choice==1:
            print("Enter the details:")
            userid=int(input("Enter User-Id: "))
            idd=int(input("Enter Task-Id: "))
            task=input("Enter task to be added in the todo-list: ")
            status=input("Enter status of the task(Yes or No): ")
            if(status=="Yes"):
                status = True
            else:
                status=False
            add_item_to_todo(userid,idd,task,status)
            print()
        elif choice==2:
            id1=int(input("Enter User-Id: "))
            display(id1)
            print()
        elif choice==3:
            id1=int(input("Enter User-Id: "))
            todo(id1)
            print()
        elif choice==4:
            print("Exiting the application......")
            break
        else:
            print("Wrong choice! Try again")
            
menu()
```
